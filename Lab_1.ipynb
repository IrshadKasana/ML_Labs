Q1 Write a python function that takes a list of 1000 integers as input and returns the prod-
uct of all the prime numbers in the list. Take care of multiplication overflow.

```python
def product_of_primes_mod64(nums):
    # Step 1: find the maximum number
    max_val = max(nums)

    # Step 2: sieve of Eratosthenes
    sieve = [True] * (max_val + 1)
    sieve[0:2] = [False, False]

    p = 2
    while p * p <= max_val:
        if sieve[p]:
            for multiple in range(p*p, max_val + 1, p):
                sieve[multiple] = False
        p += 1

    # Step 3: simulate overflow using modulo 2^64
    MOD = 2**64
    result = 1

    # Step 4: multiply primes only
    for x in nums:
        if x >= 2 and sieve[x]:
            result = (result * x) % MOD

    # Step 5: return answer
    return result
```

Q4 Sample five (10 × 1) vectors and compute their pairwise inner and outer products. Also
find out for each of the vectors, which other vector points to a similar direction as itself. bold text

```python
import numpy as np

# -----------------------------------
# Step 1: Sample five (10×1) vectors
# -----------------------------------
vectors = [np.random.randn(10, 1) for _ in range(5)]

print("Randomly sampled 5 vectors (each 10x1):")
for i, v in enumerate(vectors):
    print(f"\nVector v{i+1}:\n{v}")

# ------------------------------------------------
# Step 2: Compute all pairwise INNER products
# Inner product = vᵢᵀ vⱼ  (scalar)
# ------------------------------------------------
print("\n\nPAIRWISE INNER PRODUCTS:")
inner_products = np.zeros((5, 5))

for i in range(5):
    for j in range(5):
        inner_products[i, j] = vectors[i].T @ vectors[j]

print(inner_products)

# ------------------------------------------------
# Step 3: Compute all pairwise OUTER products
# Outer product = vᵢ vⱼᵀ  (10×10 matrix)
# ------------------------------------------------
print("\n\nPAIRWISE OUTER PRODUCTS:")
outer_products = [[None]*5 for _ in range(5)]

for i in range(5):
    for j in range(5):
        outer_products[i][j] = vectors[i] @ vectors[j].T
        print(f"\nOuter product of v{i+1} and v{j+1}:")
        print(outer_products[i][j])

# ------------------------------------------------
# Step 4: Direction similarity using cosine similarity
# Cosine similarity close to 1 → same direction
# ------------------------------------------------
def cosine_similarity(a, b):
    return float((a.T @ b) / (np.linalg.norm(a) * np.linalg.norm(b)))

print("\n\nDIRECTION SIMILARITY (Cosine Similarity):")
cos_sim = np.zeros((5, 5))

for i in range(5):
    for j in range(5):
        cos_sim[i, j] = cosine_similarity(vectors[i], vectors[j])

print(cos_sim)

# ------------------------------------------------
# Step 5: For each vector, find the most similar one
# ------------------------------------------------
print("\n\nMost similar vector for each vector:")
for i in range(5):
    sims = cos_sim[i].copy()
    sims[i] = -999  # ignore self
    most_similar = np.argmax(sims)
    print(f"Vector v{i+1} is most similar to v{most_similar+1} "
          f"(cos similarity = {sims[most_similar]:.4f})")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Standard Normal PDF formulla

def normal_pdf(x):
    return (1/np.sqrt(2*np.pi)) * np.exp(-x**2 / 2)


# Analytical derivatives

def pdf_1st(x):    # f'(x)
    return -x * normal_pdf(x)

def pdf_2nd(x):    # f''(x)
    return (x**2 - 1) * normal_pdf(x)

def pdf_3rd(x):    # f'''(x)
    return -(x**3 - 3*x) * normal_pdf(x)


# Numerical derivatives using np.gradient

x = np.linspace(-5, 5, 500)
f = normal_pdf(x)

f1_num = np.gradient(f, x)
f2_num = np.gradient(f1_num, x)
f3_num = np.gradient(f2_num, x)

# Analytical
f1 = pdf_1st(x)
f2 = pdf_2nd(x)
f3 = pdf_3rd(x)


# Plotting Visualization

plt.figure(figsize=(15, 8))

# Analytical f
plt.subplot(2, 4, 1)
plt.plot(x, f)
plt.title("Normal PDF f(x)")
plt.grid(True)

# Analytical f'
plt.subplot(2, 4, 2)
plt.plot(x, f1)
plt.title("Analytical 1st derivative f'(x)")
plt.grid(True)

# Analytical f''
plt.subplot(2, 4, 3)
plt.plot(x, f2)
plt.title("Analytical 2nd derivative f''(x)")
plt.grid(True)

# Analytical f'''
plt.subplot(2, 4, 4)
plt.plot(x, f3)
plt.title("Analytical 3rd derivative f'''(x)")
plt.grid(True)


# Numerical f'
plt.subplot(2, 4, 5)
plt.plot(x, f1_num)
plt.title("Numerical 1st derivative")
plt.grid(True)

# Numerical f''
plt.subplot(2, 4, 6)
plt.plot(x, f2_num)
plt.title("Numerical 2nd derivative")
plt.grid(True)

# Numerical f'''
plt.subplot(2, 4, 7)
plt.plot(x, f3_num)
plt.title("Numerical 3rd derivative")
plt.grid(True)

plt.tight_layout()
plt.show()
```

Q6 Implement the eigenvalue decomposition for a given 5 × 5 matrix and benchmark the
results obtained from your method against the results of numpy.linalg.eig().

```python
import numpy as np
import time


# 5×5 matrix

np.random.seed(0)
A = np.random.randn(5, 5)
print("Matrix A:\n", A)


# Eigenvalue computation using QR algorithm

def qr_eigen_decomposition(A, max_iter=1000, tol=1e-10):
    A_k = np.copy(A)
    n = A.shape[0]
    Q_total = np.eye(n)

    for i in range(max_iter):
        Q, R = np.linalg.qr(A_k)
        A_k = R @ Q
        Q_total = Q_total @ Q

        # Convergence check (off-diagonal elements small)
        off_diag = A_k - np.diag(np.diag(A_k))
        if np.linalg.norm(off_diag) < tol:
            break

    eigenvalues = np.diag(A_k)
    eigenvectors = Q_total
    return eigenvalues, eigenvectors



# Benchmark QR-based EVD

start = time.time()
eigvals_qr, eigvecs_qr = qr_eigen_decomposition(A)
end = time.time()
qr_time = end - start

print("\nEigenvalues (QR Algorithm):")
print(eigvals_qr)


# NumPy’s eigen decomposition

start = time.time()
eigvals_np, eigvecs_np = np.linalg.eig(A)
end = time.time()
np_time = end - start

print("\nEigenvalues (NumPy):")
print(eigvals_np)


# Sort eigenvalues for comparison

eigvals_qr_sorted = np.sort(eigvals_qr)
eigvals_np_sorted = np.sort(eigvals_np)


# Benchmark  numerical error

error = np.linalg.norm(eigvals_qr_sorted - eigvals_np_sorted)

print("\n===== BENCHMARK RESULTS =====")
print("QR Algorithm eigenvalues (sorted):\n", eigvals_qr_sorted)
print("NumPy eigenvalues (sorted):\n", eigvals_np_sorted)

print("\nEigenvalue Difference (L2 norm):", error)
print("QR Algorithm Time:", qr_time, "seconds")
print("NumPy eig() Time:", np_time, "seconds")
```

Q8 Write a python code to recursively read and display all files and folders for a given loca-
tion. The files should be displayed in a way that the hierarchical structure is preserved.

Filter out files that do not have one of the following extensions: jpeg, png, pdf,
doc/docx, ods, xls/xlsx, py, mp3/mp4

```python
import os

# Allowed file extensions
ALLOWED_EXT = {
    ".jpeg", ".jpg", ".png", ".pdf", ".doc", ".docx",
    ".ods", ".xls", ".xlsx", ".py", ".mp3", ".mp4"
}

filtered_files = []   # To store allowed files globally


def display_tree(path, indent=""):
    """Recursively print directory tree and collect filtered files."""
    try:
        items = sorted(os.listdir(path))
    except FileNotFoundError:
        print(" Path not found. Please check the directory.")
        return
    except PermissionError:
        print(indent + " [Permission Denied]")
        return

    for i, item in enumerate(items):
        full_path = os.path.join(path, item)
        is_last = (i == len(items) - 1)

        branch = "└── " if is_last else "├── "
        next_indent = indent + ("    " if is_last else "│   ")

        if os.path.isdir(full_path):
            print(indent + branch + f"[DIR] {item}")
            display_tree(full_path, next_indent)
        else:
            ext = os.path.splitext(item)[1].lower()
            print(indent + branch + item)

            # Collect allowed files
            if ext in ALLOWED_EXT:
                filtered_files.append(full_path)


# Ask user for directory input


print("📂 Enter the full directory path:")

location = input("\nEnter directory path: ").strip()

print(f"\n📁 Directory Tree for: {location}\n")
display_tree(location)


# Print filtered files at the end

print("\n\n FILTERED FILES (Allowed Extensions Only)\n")
if filtered_files:
    for f in filtered_files:
        print("done", f)
else:
    print("No allowed file types found.")
```

Q9 Design and implement a base class Shape that defines the methods area() and
perimeter(). Using the concept of polymorphism, create appropriate derived classes

for different geometric shapes such as rectangle, square, circle, trapezoid, and parallel-
ogram. Based on the input parameters provided, the program should dynamically de-
termine the type of shape and correctly compute its corresponding area and perimeter.

```python
import math

# Base Class

class Shape:
    def area(self):
        raise NotImplementedError("Area method not implemented")

    def perimeter(self):
        raise NotImplementedError("Perimeter method not implemented")



# Derived Classes


class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

    def perimeter(self):
        return 2 * (self.length + self.width)


class Square(Shape):
    def __init__(self, side):
        self.side = side

    def area(self):
        return self.side * self.side

    def perimeter(self):
        return 4 * self.side


class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * self.radius**2

    def perimeter(self):
        return 2 * math.pi * self.radius


class Trapezoid(Shape):
    def __init__(self, a, b, c, d, height):
        """
        a, b = parallel sides
        c, d = non-parallel sides
        h = height
        """
        self.a = a
        self.b = b
        self.c = c
        self.d = d
        self.height = height

    def area(self):
        return 0.5 * (self.a + self.b) * self.height

    def perimeter(self):
        return self.a + self.b + self.c + self.d


class Parallelogram(Shape):
    def __init__(self, base, side, height):
        self.base = base
        self.side = side
        self.height = height

    def area(self):
        return self.base * self.height

    def perimeter(self):
        return 2 * (self.base + self.side)



# Dynamic Shape Creation based on User Input

def create_shape():
    print("Choose a shape:")
    print("1. Rectangle")
    print("2. Square")
    print("3. Circle")
    print("4. Trapezoid")
    print("5. Parallelogram")

    choice = int(input("Enter choice (1-5): "))

    if choice == 1:
        l = float(input("Enter length: "))
        w = float(input("Enter width: "))
        return Rectangle(l, w)

    elif choice == 2:
        s = float(input("Enter side: "))
        return Square(s)

    elif choice == 3:
        r = float(input("Enter radius: "))
        return Circle(r)

    elif choice == 4:
        a = float(input("Enter side a (parallel): "))
        b = float(input("Enter side b (parallel): "))
        c = float(input("Enter side c: "))
        d = float(input("Enter side d: "))
        h = float(input("Enter height: "))
        return Trapezoid(a, b, c, d, h)

    elif choice == 5:
        base = float(input("Enter base: "))
        side = float(input("Enter side: "))
        height = float(input("Enter height: "))
        return Parallelogram(base, side, height)

    else:
        print("Invalid choice!")
        return None



# Main Program Execution

shape = create_shape()

if shape:
    print("\n======== RESULTS ========")
    print("Area:", shape.area())
    print("Perimeter:", shape.perimeter())
```

Q 10 Consider the following website:
https://openaccess.thecvf.com/ICCV2025?day=2025-10-23.
Write a python script to parse the source script of the above webpage and output a csv
file that lists the paper title, author names, and pdf link in adjacent columns.

```python
import requests
from bs4 import BeautifulSoup
import csv

url = "https://openaccess.thecvf.com/ICCV2025?day=2025-10-23"

html = requests.get(url).text
soup = BeautifulSoup(html, "html.parser")

rows = []

papers = soup.find_all("dt", class_="ptitle")

print("Found:", len(papers), "papers")

for dt in papers:

    # Title

    title = dt.get_text(strip=True)


    # Authors

    authors_dd = dt.find_next_sibling("dd")
    authors = authors_dd.get_text(strip=True) if authors_dd else "N/A"

   # PDF Link (Robust)

    pdf_link = None

    pdf_dd = authors_dd.find_next_sibling("dd") if authors_dd else None

    if pdf_dd:
        for a in pdf_dd.find_all("a", href=True):
            href = a["href"].lower()

            if href.endswith(".pdf") or "pdf" in a.text.lower():
                pdf_link = "https://openaccess.thecvf.com" + a["href"]
                break

    # Convert link → Clickable hyperlink formula
    if pdf_link:
        clickable = f'=HYPERLINK("{pdf_link}", "PDF")'
    else:
        clickable = "N/A"

    rows.append([title, authors, clickable])



# Save CSV

output_file = "ICCV2025_papers.csv"

with open(output_file, "w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(["Title", "Authors", "PDF Link"])
    writer.writerows(rows)

print("Saved CSV as:", output_file)
```
